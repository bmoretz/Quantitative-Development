```python
import sys
sys.path.append('../')

from matlib.enums import Color
from matlib.plotting import draw_shapes
from matlib.vectors.shapes import Polygon, Points, Segment, Arrow
from matlib.vectors.transforms import shift
```

## Dino Sample Data

```python
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]
```

```python
draw_shapes(
    Points(*dino_vectors)
)
```

```python
draw_shapes(
    Points(*dino_vectors,color=Color.Black),
    Segment((6,4), (3,1),color=Color.Blue)
)
```

```python
draw_shapes(
    Points(*dino_vectors),
    Polygon(*dino_vectors,color=Color.Blue)
)
```

```python
draw_shapes(
    Points(*[(x,x**2) for x in range(-10, 11)],color=Color.Orange),
    grid=(1,10),
    nice_aspect_ratio=False
)
```

```python
dino_vectors2 = shift(dino_vectors, k=2)
```

```python
draw_shapes(
    Points(*dino_vectors,color=Color.Blue),
    Polygon(*dino_vectors,color=Color.Blue),
    Points(*dino_vectors2,color=Color.Blue),
    Polygon(*dino_vectors2,color=Color.Blue),
)
```
