```python
import sys
sys.path.append('../')

from matlib.enums import Color
from matlib.vectors.types import Vector, make_vectors
from matlib.vectors.shapes import Polygon, Points, Segment, Arrow
from matlib.vectors.transforms import shift, scale, flip

from matlib.plotting import draw_shapes

figs_dir = '../figures/simple_vectors/'
```

## Dino Sample Data

```python
dino_vectors = make_vectors([(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)])
```

```python
draw_shapes(
    Points(*dino_vectors),
    save_as=f'{figs_dir}dino_01.png'
)
```

```python
draw_shapes(
    Points(*dino_vectors,color=Color.Black),
    Segment((6,4), (3,1),color=Color.Blue),
    save_as=f'{figs_dir}dino_02.png'
)
```

```python
draw_shapes(
    Points(*dino_vectors),
    Polygon(*dino_vectors,color=Color.Blue),
    save_as=f'{figs_dir}dino_03.png'
)
```

```python
draw_shapes(
    Points(*[Vector(x,x**2) for x in range(-10, 11)],color=Color.Orange),
    grid=(1,10),
    nice_aspect_ratio=False,
    save_as=f'{figs_dir}exp_01.png'
)
```

```python
dino_vectors2 = shift(dino_vectors, x=-1, y=-2)

draw_shapes(
    Points(*dino_vectors),
    Polygon(*dino_vectors,color=Color.Blue),
    Points(*dino_vectors2),
    Polygon(*dino_vectors2,color=Color.Red),
    save_as=f'{figs_dir}dino_04.png'
)
```

```python
dino_vectors2 = scale(dino_vectors, k=2)

draw_shapes(
    Points(*dino_vectors),
    Polygon(*dino_vectors,color=Color.Blue),
    Points(*dino_vectors2),
    Polygon(*dino_vectors2,color=Color.Red),
    save_as=f'{figs_dir}dino_05.png'
)
```

```python
draw_shapes(
    Arrow(Vector(4,3), color=Color.Green),
    save_as=f'{figs_dir}arrow_01.png'
)
```

```python
dino_vectors3 = flip(dino_vectors)

draw_shapes(
    Polygon(*dino_vectors, color=Color.Blue),
    Polygon(*dino_vectors3, color=Color.Orange),
    save_as=f'{figs_dir}dino_06.png'
)
```

### Draw 100 Dinosaurs

```python
shapes = []

y_offset = 10
for row in range(0, 10):
    
    x_offset = 2

    for col in range(0, 10):
        scaled = scale(dino_vectors, .1)
        transformed = shift(scaled, x_offset, y_offset)
        shapes += [Polygon(*transformed, color=Color.Blue)]
        x_offset = x_offset + 2

    y_offset = y_offset - 1

draw_shapes(*shapes,save_as=f'{figs_dir}dino_07.png')
```
