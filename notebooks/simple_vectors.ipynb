Suppose u = (-1, 1) and v = (1, 1) and suppose r and s are some real numbers.

Specifically, let's assume -3 < r < 3 and -1 < s < 1. Where are the possible points on the plane where the vector r*u + s*v could end up?

```python
u, v = Vector(-1, 1), Vector(1, 1)

from random import uniform

def random_r():
    return uniform(-3, 3)

def random_s():
    return uniform(-1, 1)

possibilities = []

for _ in range(500):
    r, s = random_r(), random_s()
    possibilities += [Vector((u.x * r) + (v.x * s), (u.y * r) + (v.y * s))]

draw_shapes(
    Points(*possibilities, color=Color.Blue, alpha=.5),
    save_as=f'{figs_dir}scale_01.png'
)
```
